```python
#import the required libraries for data storing ,cleaning,and visualising
#we use diferent libraries for different tasks
import pandas as pd
import numpy as np
```

```python
#Libraries for training,building of model
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Load the dataset.Here i saved the data set with the name "advertising.csv"
data=pd.read_csv('titanic.csv')
```

```python
#lets have a look on the data
data
```

```python
#since our data contains information that we not need for our analysis we clean it by dropping some of the non important columns
new_data=data.drop(['PassengerId', 'Name', 'Ticket', 'Cabin','Embarked'],axis=1)
```

```python
#lets have a look on the new data
new_data
```

```python
#in this step we change string data type to numerical data by making the values female represented with 1 and men with 0
new_data['Sex'] = new_data['Sex'].map({'female': 1, 'male': 0})
```

```python
#in age column we find some NaN values which shouldn't be there so we fill median value of age in that missing places
new_data['Age'].fillna(new_data['Age'].median(), inplace=True)
```

```python
# we follow same step for Fare column too
new_data['Fare'].fillna(new_data['Fare'].median(), inplace=True)
```

```python
#Now we split the data into two parts one as a target data('y') and other as variables('x')
X = new_data.drop('Survived', axis=1)
y = new_data['Survived']
```

```python
# In this step we split the data into testing data and training data with test size=0.2 ie 20% data for testing
#and 80% training
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# we select our suitable model for training and testing of our data
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
#here we try to predit the data
y_pred = model.predict(X_test)
```

```python
# we check the accuracy of data between predicted and tested data
accuracy = accuracy_score(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)
```

```python
print(f'Accuracy: {accuracy:.2f}')
print('Classification Report:\n', classification_rep)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
# in this step we calculate the mean square error
rsm_error = np.sqrt(mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', rsm_error)
```
